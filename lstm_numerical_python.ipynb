```python
import math

def sigmoid(x):
    return 1.0 / (1.0 + math.exp(-x))

def tanh(x):
    return math.tanh(x)

print(f"sigmoid(0) = {sigmoid(0)}")
print(f"tanh(0)    = {tanh(0)}")
```

```python

Wf, Wi, Wg, Wo = 0.5, 0.6, 0.4, 0.7

Uf, Ui, Ug, Uo = 0.1, 0.2, 0.3, 0.4

bf, bi, bg, bo = 0.1, 0.2, 0.3, 0.1


Wy, by = 1.5, 0.5

h0, c0 = 0.0, 0.0

inputs = [1, 2, 3, 4]

print(f"Input sequence : {inputs}")
print(f"h0 = {h0},  c0 = {c0}")
```

```python
def lstm_step(x, h_prev, c_prev):
    f   = sigmoid(Wf * x + Uf * h_prev + bf)
    i   = sigmoid(Wi * x + Ui * h_prev + bi)
    g   =    tanh(Wg * x + Ug * h_prev + bg)
    o   = sigmoid(Wo * x + Uo * h_prev + bo)
    c_t = f * c_prev + i * g
    h_t = o * tanh(c_t)
    return h_t, c_t, {"f": f, "i": i, "g": g, "o": o}

print("lstm_step() is ready.")
```

```python
h, c = h0, c0

h, c, gates = lstm_step(inputs[0], h, c)

print("Time Step t=1,  x=1")
print(f"  Forget gate  f_1 = {gates['f']:.4f}")
print(f"  Input  gate  i_1 = {gates['i']:.4f}")
print(f"  Candidate    g_1 = {gates['g']:.4f}")
print(f"  Output gate  o_1 = {gates['o']:.4f}")
print(f"  Cell  state  c_1 = {c:.4f}")
print(f"  Hidden state h_1 = {h:.4f}")
```
